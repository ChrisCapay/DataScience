```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df_data_NYC = pd.read_csv('311_Service_Requests.csv')
df_data_NYC.head()
```

```python
#Rename a single column name
df_data_NYC.rename(columns={'Unnamed: 0':'CompCount'},inplace=True)
# replacing blank spaces in all column names with '_'  
df_data_NYC.columns =[column.replace(" ", "_") for column in df_data_NYC.columns]
#list all column names
df_data_NYC.columns
```

```python
df_data_BX = pd.read_csv('BX_18v1.csv')
df_data_BX.head()
```

```python
#create a new df for Bronx with the key features

df_data_BX= df_data_BX[['Address','BldgArea','BldgDepth','BuiltFAR','CommFAR','FacilFAR','Lot','LotArea','LotDepth',
               'NumBldgs','NumFloors','OfficeArea','ResArea','ResidFAR','RetailArea','YearBuilt','YearAlter1','ZipCode','YCoord','XCoord']]
df_data_BX.head()
```

```python
#limit rows to Bronx Complaints
df_data_NYC.query('Borough =="BRONX"', inplace=True)
df_data_NYC.head()
```

```python
#Add a compindex ; true if heat related
#takes a nested sequence of objects or numpy arrays as inputs and returns a single numpy array or a tuple of numpy arrays

def conditions(x):
    if x == "HEAT/HOT WATER":
        return 1
    elif x == "HEATING":
        return 1
    else:
        return 0

func = np.vectorize(conditions) 
complaint_class = func(df_data_NYC["Complaint_Type"])

df_data_NYC["compindex"] = complaint_class
df_data_NYC.head()
```

```python
#drop rows in Zip with NaN
df_data_NYC.dropna(subset=['Incident_Zip'],inplace=True)
df_data_BX.dropna(subset=['ZipCode'],inplace=True)
```

```python
#join NYC and Bronx using Address as pk
df_merged_BX = pd.merge(df_data_NYC, df_data_BX, left_on='Incident_Address', right_on='Address')
df_merged_BX.head()
```

```python
df_merged_BX.shape
```

```python
#create dataframe for features
BX_Heat = df_merged_BX[['BldgArea', 'BldgDepth', 'BuiltFAR', 'CommFAR', 'FacilFAR', 'Lot',
       'LotArea', 'LotDepth', 'NumBldgs', 'NumFloors', 'ResArea',
       'ResidFAR', 'RetailArea', 'YearBuilt', 'YearAlter1', 'compindex']]
BX_Heat.head()
```

```python
BX_Heat.fillna(BX_Heat.mean(), inplace=True)
BX_Heat.isnull().sum() 
```

```python
y = BX_Heat['compindex']
X = BX_Heat.drop(['compindex'], axis=1)
```

```python
#Normalize data
from sklearn.preprocessing import StandardScaler
X= StandardScaler().fit_transform(X)
X[0:5]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Validation set:', X_val.shape,  y_val.shape)
```

```python
#KNN

from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

#Train Model and Predict  
neigh = KNeighborsClassifier(n_neighbors = 5).fit(X_train,y_train)
yhat = neigh.predict(X_val)
print("KNN Accuracy at k = 5: ", metrics.accuracy_score(y_val, yhat))
```

```python
#Decision Tree

from sklearn.tree import DecisionTreeClassifier

BXTree = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
BXTree # it shows the default parameters
BXTree.fit(X_train,y_train)

predTree = BXTree.predict(X_val)
print (predTree [0:5])
print (y_val [0:5])

from sklearn import metrics
print("DecisionTrees's Accuracy at max_depth = 4: ", metrics.accuracy_score(y_val, predTree))
```

```python
#SVM

from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report, confusion_matrix

LSVC = LinearSVC()
LSVC.fit(X_train, y_train) 
yhat = LSVC.predict(X_val)
print("SVC Accuracy: ", metrics.accuracy_score(y_val, yhat))

#print("Radial Basis Function Accuracy: ", jaccard_similarity_score(y_val, yhat))
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import jaccard_similarity_score

LR = LogisticRegression(C=0.01, solver='liblinear').fit(X_train,y_train)
yhat = LR.predict(X_val)
#print("Radial Basis Function Accuracy: ", jaccard_similarity_score(y_val, yhat))
print("LR Accuracy at: ", metrics.accuracy_score(y_val, yhat))
```

```python
#Evaluate Model using test set

from sklearn.metrics import f1_score
# for f1_score please set the average parameter to 'micro'
from sklearn.metrics import log_loss
from sklearn.metrics import jaccard_similarity_score

def jaccard_index(predictions, true):
    if (len(predictions) == len(true)):
        intersect = 0;
        for x,y in zip(predictions, true):
            if (x == y):
                intersect += 1
        return intersect / (len(predictions) + len(true) - intersect)
    else:
        return -1
    
test_y = BX_Heat['compindex'].values
test_y[0:5]
```

```python
neigh = KNeighborsClassifier(n_neighbors = 5).fit(X_val,y_val)
yhat = neigh.predict(X_val)

print("F1 score:",f1_score(y_val, yhat, average='micro'))
print("Jaccard Index:",jaccard_index(yhat, y_val))
```
