```python
import pandas as pd
import numpy as np
import datetime as dt

df_2005_to_2008 = pd.read_csv('C:\\Users\\user\\Nasdaq\\dwa_bambu_historical_05212005_to_05212008')
df_2008_to_2011 = pd.read_csv('C:\\Users\\user\\Nasdaq\\dwa_bambu_historical_05222008_to_05222011')
df_2011_to_2014 = pd.read_csv('C:\\Users\\user\\Nasdaq\\dwa_bambu_historical_05232011_to_05232014')
df_2014_to_2017 = pd.read_csv('C:\\Users\\user\\Nasdaq\\dwa_bambu_historical_05242014_to_05242017')
df_2017_to_2020 = pd.read_csv('C:\\Users\\user\\Nasdaq\\dwa_bambu_historical_05252017_to_05252020')
df_tickers = pd.read_csv('C:\\Users\\user\\Nasdaq\\isins.csv')
df_prices = pd.read_csv('C:\\Users\\user\\Nasdaq\\historicalprices.csv')
#df_dow = pd.read_csv('C:\\Users\\user\\Nasdaq\\dow.csv') 
```

```python
print(df_prices.shape)
```

```python
df_prices.rename(columns={"RIC": "date"},inplace=True)
```

```python
df_prices.set_index('date',inplace=True)
```

```python
df_prices_transposed = df_prices.transpose()

```

```python
#convert index to datetime so that you can get the weekday name.price_dates.dt.weekday_name
price_dates = pd.to_datetime(df_prices_transposed.index.values).to_series()
df_prices_transposed['day_of_week'] = price_dates.dt.weekday_name
```

```python
lastprice_friday = df_prices_transposed.query('day_of_week =="Friday"')
```

```python
lastprice_friday.drop(['day_of_week'], axis=1, inplace=True)
```

```python
lastprice_friday_numeric = lastprice_friday.apply(pd.to_numeric, errors='coerce')
```

```python
df_4w = lastprice_friday_numeric.pct_change(4)
df_12w = lastprice_friday_numeric.pct_change(12) 
df_26w = lastprice_friday_numeric.pct_change(26)
df_52w = lastprice_friday_numeric.pct_change(52)
```

```python
lastprice_friday_stacked = lastprice_friday_numeric.stack().reset_index()
lastprice_friday_stacked.rename(columns={"level_0": "date", "date": "ticker", 0:"close"},inplace=True)
lastprice_friday_stacked.set_index('date',inplace=True)
```

```python
#create a stacked df for 4w,12w,162,52w

df_4w_stacked = df_4w.stack().reset_index()
df_4w_stacked.rename(columns={"level_0": "date", "date": "ticker", 0:"4w_change"},inplace=True)
df_4w_stacked.set_index('date',inplace=True)

df_12w_stacked = df_12w.stack().reset_index()
df_12w_stacked.rename(columns={"level_0": "date", "date": "ticker", 0:"12w_change"},inplace=True)
df_12w_stacked.set_index('date',inplace=True)

df_26w_stacked = df_26w.stack().reset_index()
df_26w_stacked.rename(columns={"level_0": "date", "date": "ticker", 0:"26w_change"},inplace=True)
df_26w_stacked.set_index('date',inplace=True)

df_52w_stacked = df_52w.stack().reset_index()
df_52w_stacked.rename(columns={"level_0": "date", "date": "ticker", 0:"52w_change"},inplace=True)
df_52w_stacked.set_index('date',inplace=True)
```

```python
lastprice_stacked_all = pd.merge(lastprice_friday_stacked, df_4w_stacked,left_on=['date','ticker'], right_on=['date','ticker'], how="left")
lastprice_stacked_all = pd.merge(lastprice_stacked_all, df_12w_stacked,left_on=['date','ticker'], right_on=['date','ticker'], how="left")
lastprice_stacked_all = pd.merge(lastprice_stacked_all, df_26w_stacked,left_on=['date','ticker'], right_on=['date','ticker'], how="left")
lastprice_stacked_all = pd.merge(lastprice_stacked_all, df_52w_stacked,left_on=['date','ticker'], right_on=['date','ticker'], how="left")

```

```python
#add RIC to the DWA files.

df_2005_to_2008 = pd.merge(df_2005_to_2008, df_tickers, left_on='ticker', right_on='ticker', how="left")
df_2008_to_2011 = pd.merge(df_2008_to_2011, df_tickers, left_on='ticker', right_on='ticker', how="left")
df_2011_to_2014 = pd.merge(df_2011_to_2014, df_tickers, left_on='ticker', right_on='ticker', how="left")
df_2014_to_2017 = pd.merge(df_2014_to_2017, df_tickers, left_on='ticker', right_on='ticker', how="left")
df_2017_to_2020 = pd.merge(df_2017_to_2020, df_tickers, left_on='ticker', right_on='ticker', how="left")
```

```python
#add weekly percentage changes to each of the yearly files

df_2008_to_2011_with_prices = pd.merge(df_2008_to_2011, lastprice_stacked_all, left_on=['RIC','date'], right_on=['ticker','date'])
df_2011_to_2014_with_prices = pd.merge(df_2011_to_2014, lastprice_stacked_all, left_on=['RIC','date'], right_on=['ticker','date'])
df_2014_to_2017_with_prices = pd.merge(df_2014_to_2017, lastprice_stacked_all, left_on=['RIC','date'], right_on=['ticker','date'])
df_2017_to_2020_with_prices = pd.merge(df_2017_to_2020, lastprice_stacked_all, left_on=['RIC','date'], right_on=['ticker','date'])
```

```python
df_2008_to_2011_with_prices.shape
```

```python
df_final_nasdaq = pd.concat([df_2008_to_2011_with_prices,df_2011_to_2014_with_prices,
                       df_2014_to_2017_with_prices,df_2017_to_2020_with_prices])
```
